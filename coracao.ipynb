```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2,2,100)
```

```python
y1 = np.sqrt((1-(np.abs(x)-1)**2))
y2 = np.arccos(1-np.abs(x)) - np.pi
```

```python
plt.plot(x,y1,'r')
plt.plot(x,y2,'r')
plt.show()
```
